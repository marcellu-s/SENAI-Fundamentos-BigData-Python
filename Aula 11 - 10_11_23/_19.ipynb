```python
# 1) Importar as bibliotecas Pandas, Seaborn e Matplotlib;
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2) Carregue o arquivo dados-satisfação.xlsx;
dataframe = pd.read_excel('./archives/dados-satisfação.xlsx')
```

```python
dataframe.head()
```

```python
'''
Realizar o tratamento dos dados categóricos, converta-os para dados
numéricos. As colunas que terão os dados convertidos serão: Elogio quanto ao
app, Reclamação quanto ao app, Elogio a Instituição, Reclamação a Instituição
e Não Classificável.
Nestas colunas, você deve substituir NÃO para 0 e SIM para 1. Converta os
dados para o tipo INT.
'''
columns = ['Elogio quanto ao app', 'Reclamação quanto ao app', 'Elogio a Instituição', 'Reclamação a Instituição', 'Não Classificável']

for column in columns:

    dataframe[column] = dataframe[column].replace(['SIM', 'NÃO', 'NAO'], [1, 0, 0])

dataframe.head()
```

```python
# 4) Faça as seguintes substituições na coluna classificação:
dataframe['Classificação'] = dataframe['Classificação'].replace([5, 4, 3, 2, 1], [1, 1, 0, -1, -1])

dataframe.head()
```

```python
'''
5) Crie um dataframe chamado df_classificacao, para agrupar os dados por
instituição, calculando a média da da coluna Classificação.
'''
dataframe_classification = dataframe.groupby(by='Instituição')['Classificação'].agg('mean').reset_index()
```

```python
# 6) Multiplique os valores da coluna Classificação por 100.
dataframe_classification['Classificação'] = (dataframe_classification['Classificação'] * 100).round(1)
```

```python
# 7) Carregue a paleta de cores Accent, com o comando sns.set_palette;
sns.set_palette(palette='Accent')


# 8) Carregue o estilo darkgrid, utilizando o comando sns.set_style;
sns.set_style(style='darkgrid')
```

```python
'''
9) Gere um gráfico de colunas para apresentar a classificação média por instituição,
e escreva um texto com sua análise do NPS.
'''
sns.barplot(x=dataframe_classification['Instituição'], y=dataframe_classification['Classificação'])

plt.title('Classificação das instituições bancárias')

plt.xticks(rotation=30)

plt.show()
```

```python
'''
10) Repita os passos 5, 6 e 9, considerando a coluna Elogio quanto ao app e
Reclamação quanto ao App;
'''

output_graphic = './graphics/'

columns = ['Elogio quanto ao app', 'Reclamação quanto ao app', 'Elogio a Instituição', 'Reclamação a Instituição']

for column in columns:

    dataframe_group = dataframe.groupby(by='Instituição')[column].agg('mean').reset_index()

    dataframe_group[column] = (dataframe_group[column] * 100).round(1)

    plt.figure(figsize=(10, 5))

    sns.barplot(x=dataframe_group['Instituição'], y=dataframe_group[column])

    plt.title(f'{column}')

    plt.savefig(f'{output_graphic}{column}.jpg')
```
