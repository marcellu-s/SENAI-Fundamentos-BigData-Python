```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rand
import pandas as pd
```

### Múltiplas curvas em um gráfico

```python
x = np.linspace(0, 2 * np.pi, 100) # Gerar 100 números entre 0 e 2*PI
y = np.sin(x)
z = np.cos(x)

plt.title('BATATA')

plt.plot(x, y, label='Seno', color='#1b4965')
plt.plot(x, z, label='Cosseno', color='#ccdbdc')

plt.legend(loc='lower left')

plt.xlabel('Arroz')
plt.ylabel('Salada')

plt.savefig('./graphics/funcoes-trigonometricas.jpg')

plt.show()
```

### Múltiplos subgráficos

```python
# HORIZONTAL

# nrows - Número de linhas; ncols - Número de colunas; pos - Posíção do gráfico

# Primeiro gráfico
plt.subplot(2, 1, 1)

plt.plot(x, y)

# Segundo gráfico
plt.subplot(2, 1, 2)

plt.plot(x, z)

plt.savefig('./graphics/grafico-horizontal.jpg')

plt.show()
```

```python
# VERTICAL

plt.subplot(1, 2, 1)

plt.plot(x, y)

plt.subplot(1, 2, 2)

plt.plot(x, z)

plt.savefig('./graphics/grafico-vertical.jpg')

plt.show()
```

### Gráfico de pizza

```python
# Dados
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Distanciamento das fatias
explode = [0.1, 0, 0, 0, 0, 0]

# Gerar o gráfico
plt.pie(popularity, explode=explode, labels=languages, autopct='%1.1f%%', shadow=True)

# labels -> Rótulos da categoria
# autopct -> Rótulos numéricos
# %1.1f%% -> Formatar os valores dos rótulos numéricos com 1 casa após a virgula
# shadow -> Sombra

plt.show()
```

### Gráfico de dispersão

> O gráfico de dispersão pode ser usado com
vários para analisar a correlação entre
variáveis.

```python
# Gerando 200 dados aleatórios
x = rand(100)
y = rand(100)

# Criar o gráfico de dispersão
sns.scatterplot(x=x, y=y)
```

### Gerando um gráfico de dispersão a partir de uma base de dados

```python
dataframe = pd.read_csv('./archives/sea-level.csv')

dataframe = dataframe.dropna()
```

```python
years = dataframe.Year.drop_duplicates().tolist()
nivel = dataframe['CSIRO - Adjusted sea level (inches)'].tolist()

sns.scatterplot(x=years, y=nivel)

plt.grid()

plt.show()
```
