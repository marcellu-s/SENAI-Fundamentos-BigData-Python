```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('./archives/company_sales_data.csv')
```

```python
dataframe.head()
```

### Exercício 04 

> Leia os dados de vendas de pasta de dente de cada mês e mostre-os
usando um gráfico de dispersão. Além disso, adicione uma grade no gráfico. o estilo da linha de grade deve “–“.

```python
dataframe_group = dataframe.groupby(by=['month_number'])['toothpaste'].sum()

months = dataframe_group.index

toothpaste_values = dataframe_group.values

plt.scatter(months, toothpaste_values, label='Tooth paste Sales Data')

plt.title('Tooth paste Sales Data')

plt.xlabel('Month number')
plt.ylabel('Number of units Sold')

plt.xticks(months)

plt.grid(ls='--')

plt.legend(loc='upper left')

plt.show()
```

### Exercício 05

> Leia os dados de vendas de cremes faciais e produtos de lavagem facial
e mostre-os usando o gráfico de barras.
O gráfico de barras deve exibir o número de unidades vendidas por mês para cada
produto. Adicione uma barra separada para cada produto no mesmo gráfico.

```python
# Agrupando os dados necessários
dataframe_group = dataframe.groupby('month_number')[['facecream', 'facewash']].sum()

# Definindo os meses
months = dataframe_group.index

# Definindo os dados
facecream_per_month = dataframe_group.facecream.values
facewash_per_month = dataframe_group.facewash.values

# Largura da barra
bar_width = 0.25

# Definindo as posições das barras
pos = [pos - 0.125 for pos in months]
pos_2 = [pos + 0.125 for pos in months]

# Redimensionando o tamanho do gráfico
plt.figure(figsize=(10, 5))

# # Gerando o gráfico
plt.bar(pos, facecream_per_month, width=bar_width, label='Face Cream sales data')
plt.bar(pos_2, facewash_per_month, width=bar_width, label='Face wash sales data')

plt.xticks(months)

plt.title('Faceawash and facecream sales data')

plt.xlabel('Month number')
plt.ylabel('Sales units in number')

plt.grid(ls='--', lw=1)

plt.legend(loc='upper left')

plt.show()
```

### Exercício 6 

> Leia os dados de vendas de sabonetes de todos os meses e mostre-os
em um gráfico de barras. Salve este gráfico em um diretório.

```python
dataframe_group = dataframe.groupby('month_number')['bathingsoap'].sum()

months = dataframe_group.index
values = dataframe_group.values

sns.barplot(x=months, y=values)

plt.title('Bathingsoap sales data')

plt.xlabel('Month number')
plt.ylabel('Sales units in number')

plt.grid()

plt.savefig('./graphics/sabonetes-por-mes.jpg')

plt.show()
```

### Exercício 07

> Leia o lucro total de cada mês e mostre-o usando o histograma para ver
as faixas de lucro mais comuns.

```python
dataframe_group = dataframe.groupby('month_number')['total_profit'].sum()

sns.histplot(dataframe_group)

plt.title('Profit data')

plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
```
