```python
import matplotlib.pyplot as plt
import numpy as np
```

### Primeiros passos com Matplotlib

```python
# Criando os arrays (matrizes) para utlizar no gráfico
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

### Gerando o gráfico - DEFAULT


```python
plt.plot(x, y)
```

### Gerando o gráfico - Colocando o marcador e cor

```python
plt.plot(x, y, marker='.', c='purple')
```

### Gerando o gráfico - Modificando a forma de ligação e esilo da linha

```python
plt.plot(x, y, marker='.', c='purple', ds='steps', ls='--')

# marker = Marcador
# c = Cor da linha
# ds = ?
# ls = Estilo da linha
```

### Alterando as cores dos marcadores

```python
plt.plot(x, y, marker='.', c='purple', mec='red')
```

### Adicionando outros elementos

```python
# Adicionando um título ao gráfico
plt.title('Título do gráfico', fontsize=15, c='purple')

# Label do eixo X
plt.xlabel('Valores de X', fontsize=12)

# Label do eixo Y
plt.ylabel('Valores de Y', fontsize=12)

# Estilo dos valores do eixo X
plt.xticks(fontsize = 8, c ='blue')

# Estilo dos valores do eixo Y
plt.yticks(fontsize = 8, c ='orange')

plt.plot(x,y)
```
