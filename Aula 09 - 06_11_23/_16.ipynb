```python
# Importar as bibliotecas Pandas, Matplotlib, Seaborn e OS.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar os dados do arquivo DadosVendaCarros.xlsx;
dataframe = pd.read_excel('./archives/DadosVendaCarros.xlsx')
```

```python
# Visualizar o cabeçalho do dataframe
dataframe.head()
```

```python
# Renomear as colunas
dataframe.columns = [
    'DataNotaFiscal', 'Fabricante', 'Estado', 'ValorVenda', 'ValorCusto',
    'TotalDesconto', 'CustoEntrega', 'CustoMaoDeObra', 'NomeCliente',
    'Modelo', 'Cor', 'Ano', 'Mês'
]

dataframe.head()
```

```python
# Verificar de há dados nulos ou faltantes.
dataframe.isnull().sum()
```

```python
# Remover linhas de dados faltantes
dataframe = dataframe.dropna()

dataframe.isnull().sum()
```

```python
# Criar um gráfico de colunas para representar o total vendido por ano;

dataframe_group = dataframe.groupby(by=['Ano'])['ValorVenda'].sum()

# Eixo X

years = []
value_per_year = []

for i, year in enumerate(dataframe_group.index):

    years.append(int(year))

    value_per_year.append(int(dataframe_group.values[i]))

sorted(value_per_year)

plt.xticks(years)

plt.bar(years, value_per_year)

plt.show()
```
