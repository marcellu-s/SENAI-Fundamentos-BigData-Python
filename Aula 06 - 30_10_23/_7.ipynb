### Importando a bilbioteca PANDAS

```python
import pandas as pd
```

### PANDAS SERIES

> Uma série em Pandas é como uma coluna de uma tabela.
É uma matriz unidimensional que contém dados de qualquer tipo. Podemos usar
a função pd.Series() para criar uma série

```python
# Criando um pandas Series a partir de uma lista
ages = pd.Series([10, 20, 30, 40, 50, 60])

print(type(ages))

print(ages)
```

### DATAFRAME

> Um Pandas DataFrame é uma estrutura de dados bidimensional, como um array
bidimensional ou uma tabela com linhas e colunas.

```python
# Caminho do arquivo
document_path = "./bank-full.xlsx"

# Carregar o arquivo para o DATAFRAME
dataframe = pd.read_excel(document_path, sheet_name="Sheet 1 - bank-full")
```

### Renomear as colunas

> Modificar TODAS as colunas

Requisitos: 
  * As colunas devem estar na mesma ordem dos nomes
  * A quantidade de nomes devem ser a mesma que a de colunas

```python
dataframe.columns = [
    'idade',
    'job',
    'estado_civil',
    'educação',
    'default',
    'balance',
    'housing',
    'empréstimo',
    'contato',
    'dia',
    'mês',
    'duração',
    'campanha',
    'pdays',
    'previous',
    'poutcome'
]

dataframe.head(2)
```

> Modificar colunas específicas

```python
dataframe = dataframe.rename(columns={
    'job': 'profissão'
})

dataframe.head(1)
```
