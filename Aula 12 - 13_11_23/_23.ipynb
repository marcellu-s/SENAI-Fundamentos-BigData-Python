### Situação Problema: Previsão de Vendas de Sorvetes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
ice_cream_dataframe = pd.read_csv('./archives/IceCreamData.csv')

# Detalhes geral 
ice_cream_dataframe.info()
```

### Gerar visualizações gráficas

```python
sns.jointplot(data=ice_cream_dataframe, x='Temperature', y='Revenue', color='red')
```

```python
# Gráfico de dispersão e linha de tendência
sns.lmplot(data=ice_cream_dataframe, x='Temperature', y='Revenue')
```

```python
# Variável independente (X) -> Entrada
x = ice_cream_dataframe[['Temperature']].values

# Variável dependete (Y) -> Saída
y = ice_cream_dataframe[['Revenue']].values

# Separar dados de treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, train_size=0.75, random_state=42)
```

```python
# Criação do modelo
model = LinearRegression()
```

```python
# Treinamento
model.fit(x_train, y_train)
```
